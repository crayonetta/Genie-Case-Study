# Scen 2 Task 2 
Generate HP ageing buckets

Task 2: Generate hp aging bucket by number of account and general ledger balance snapshot based on Genie Dataset 1 and Genie Dataset 4. 
Aging buckets refer to category M0 : 0 days, category M1 : 1 – 30 days, category M2 : 31– 60 days, category M3 : 61 – 90 days, category > M3 : > 90 days. 
Aging is defined based on days different between reporting day and contract commencement date. Candidate could provide the result in python format or using any BI Tools.

HP refers to hire purchase: difference between reporting date and contract commencement date [Agrt Date, Date Settled] and [Agrt Date, Last paid date]
General Ledger: GL or GL Bal

```python
print(df1.info())
print(df1.head)

print(df4.info())
print(df4.head)
```

```python
# merge df1 and df4
task2_df = pd.merge(df1, df4, left_on='Agrt No.', right_on='Agrt No', how='outer')
```

```python
print(task2_df.columns)
```

```python
# difference in days create column called aging
# for df1 columns
task2_df['Ageing'] = (task2_df['Date Settled'] - task2_df['Agrt Date_x']).dt.days
# for df2 columns
task2_df['Ageing'] = (task2_df['Last Paid Date'] - task2_df['Agrt Date_y']).dt.days
```

```python
print(task2_df.head)
```

```python
# Combine Agrt No columns, filling in missing values from one column with values from another
task2_df['Agrt_No_combined'] = task2_df['Agrt No.'].combine_first(task2_df['Agrt No'])
```

```python
# General Ledger columns
#replace NaN values in GL and GL Bal
task2_df['GL Bal'] = task2_df['GL Bal'].fillna(0)
task2_df['GL '] = task2_df['GL '].fillna(0)
# combine different GL and GL Bal columns
task2_df['total_GL_balance'] = task2_df['GL Bal'] + task2_df['GL ']
```

```python
# Define age buckets based on the column ageing
def ageing_catergory(days):
    if days == 0:
        return 'M0'
    elif 1 <= days <= 30:
        return 'M1'
    elif 31 <= days <= 60:
        return 'M2'
    elif 61 <= days <= 90:
        return 'M3'
    else:
        return '>M3'
```

```python
# Creating a new column for 'Aging 
task2_df['Ageing Categories'] = task2_df['Ageing'].apply(ageing_catergory)
```

```python
# creating table
# group by aging bucket, count unique accounts and sum the general ledger balance
ageing_summary = task2_df.groupby('Ageing Categories').agg(
    num_accounts=('Agrt_No_combined', 'nunique'),  # use 'nunique' to count unique account numbers
    total_gl_balance=('total_GL_balance', 'sum')   # sum of general ledger 
).reset_index()

print(ageing_summary)
```
